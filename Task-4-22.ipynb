По окружности длиной 60 м. равномерно в одном направлении движутся 2 точки, одна из которых делает
полный оборот на 5 с. быстрее другой. При этом совпадение точек происходит через 1 мин. Определить
скорости точек в м/с. В ответ записать их сумму.

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

sol = solve(Eq(60 / (60 / x - 60 / (x + 5)),60), [x])
print("time:",[sol[1],sol[1] + 5])
v1 = 60 / sol[1]
v2 = 60 / (sol[1] + 5)
v = [v1, v2]
print("Velocity:",v)
print("Sum of velocities:", (v1 + v2))
```

```python
from sympy.abc import x, y
from sympy import Eq
from sympy import solve

sol = solve([Eq(x-y, -5), Eq(60 / (60 / x - 60 / y), 60)], [x,y])
print("time:", sol[1])
v1 = 60 / sol[1][0]
v2 = 60 / sol[1][1]
v = [v1, v2]
print("Velocity:",v)
print("Sum of velocities:", (v1 + v2))
```
